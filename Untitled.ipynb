```python
from PIL import Image
from torchvision import transforms
import numpy as np
import matplotlib.pyplot as plt
import torchvision
```

```python
img = Image.open("/home/pml02/datasets/ImageNet_train_64x64/n04347754/n04347754_99832.png").convert("RGB")
```

```python
print(img.size)
print(img.mode)
```

```python
transformation = transforms.Compose([
                transforms.ToTensor(),                 # [0,1]
                transforms.Normalize([0.5]*3, [0.5]*3) # -> [-1,1]
            ])
```

```python
img_arr = np.array(img)
print(img_arr.shape)
```

```python
new_img = transformation(img)
print(new_img.shape)
print(new_img)
```

```python
plt.imshow(img)
plt.axis('off')
plt.show()  # sačuva u fajl
plt.close()
```

```python
nimg = (new_img + 1) / 2

# Ako je tensor u obliku [C, H, W], matplotlib očekuje [H, W, C]
perimg = nimg.permute(1, 2, 0).cpu().detach()

plt.imshow(perimg)
plt.axis('off')  # uklanja ose
plt.show()
```
