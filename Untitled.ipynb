```python
from PIL import Image
from torchvision import transforms
import numpy as np
import matplotlib.pyplot as plt
import torchvision
import os
```

```python
img = Image.open("/home/pml02/datasets/ImageNet_train_original/n04347754/n04347754_99832.jpg").convert("RGB")
```

```python
print(img.size)
print(img.mode)
```

```python
transformation = transforms.Compose([
                transforms.ToTensor(),                 # [0,1]
                transforms.Normalize([0.5]*3, [0.5]*3) # -> [-1,1]
            ])
```

```python
img_arr = np.array(img)
print(img_arr.shape)
```

```python
new_img = transformation(img)
print(new_img.shape)
print(new_img)
```

```python
plt.imshow(img)
plt.axis('off')
plt.show()  # sauva u fajl
plt.close()
```

```python
nimg = (new_img + 1) / 2

# Ako je tensor u obliku [C, H, W], matplotlib oekuje [H, W, C]
perimg = nimg.permute(1, 2, 0).cpu().detach()

plt.imshow(perimg)
plt.axis('off')  # uklanja ose
plt.show()
```

```python
def sample_image_noise(batch_size = 128, channels=3, height=64, width=64):
    return np.random.randn(batch_size, channels, height, width).astype(np.float32)
```

```python
noise = sample_image_noise(1)[0]
```

```python
print(noise.shape)
```

```python
# 1. Transponuj dimenzije u (H, W, C)
img = np.transpose(noise, (1, 2, 0))

# 2. Skaliranje u [0,1] (ako vrednosti nisu ve u tom opsegu)
img = (img - img.min()) / (img.max() - img.min())

# 3. Prikaz slike
plt.imshow(img)
plt.axis('off')
plt.show()
```
